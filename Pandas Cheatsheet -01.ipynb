### Filtering Data -> Conditions and Masking using boolean Series

```python
# Based on Conditions
# also can create a mask - >  mask = data["city"] == "Hyderabad"
data["city"] == "Hyderabad"
```

```python
data[data["city"] == "Hyderabad"]
```

```python
data[data["city"] == "Hyderabad"].shape
```

```python
def get_Match_Count_City(city):
    return data[data["city"] == city].shape[0]
get_Match_Count_City("Pune") 
```

```python
data["city"].unique()
```

```python
get_Match_Count_City("Dharamsala") 
```

```python
# Working with TWO Conditions
condition_01 = data["city"] == "Hyderabad"
condition_02 = data["date"] > "2010-04-08"

data[condition_01 & condition_02]
```
