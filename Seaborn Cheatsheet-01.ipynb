## Relational/Statistical Plots

### relplot (Figure level Function)

```python
plt.style.use("fivethirtyeight") 
```

```python
data= sns.load_dataset('tips') # Data available from Seaborn
```

```python
data.head()
```

#### Scatter plot between total_bill and tip using RELPLOT

```python
sns.relplot(x= "total_bill",y="tip",data = data)
plt.show()
```

The plot shows an Linear Relationship between the total_bill and tips

#### Hue parameter

```python
sns.relplot(x= "total_bill",y="tip",hue="sex", data = data)
plt.show()
```

```python
sns.relplot(x= "total_bill",y="tip",hue="smoker", data = data)
plt.show()
```

#### Style Parameter

```python
sns.relplot(x= "total_bill",y="tip",hue="smoker", style ="sex", data = data)
plt.show()
```

#### Size Parameter

```python
sns.relplot(x= "total_bill",y="tip",hue="smoker", style ="sex",size ="size", sizes = (10,100)  ,data = data)
plt.show()
```

Hue parameter & Style Parameter uses - Categorical Data :::::
Size paramter - Continous or Numnerical Data

## Categorical Scatterplots - catplot with kind - Strip and Swarm

```python
 # X-axis ---> Categorical Data
 # Y-axis ---> Numerical Data 
```

```python
plt.style.use("fivethirtyeight")
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.catplot(x="day",y="tip",kind="strip",jitter =0.02,data =tips) # STRIP PLOT
# Jitter adds noise/variation
```

```python
sns.catplot(x="day",y="tip",kind="swarm",data =tips) # SWARM PLOT , Distribution of Data
```

```python
sns.catplot(x="day",y="tip",kind="swarm",hue = "sex",data =tips)
```

```python
sns.swarmplot(x="day",y="tip",hue = "smoker",data =tips)
```

```python
sns.stripplot(x="day",y="tip",hue = "smoker",data =tips)
```

### DIST PLOTS

#Distribution Plots 

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic["age"].fillna(titanic["age"].mean(), inplace=True)
```

```python
titanic.isnull().sum()
```

```python
sns.distplot(titanic["age"])
```

```python
# Plotting only Histogram
sns.distplot(titanic["age"],bins=10,kde=False)
```

```python
# Plotting only KDE -> Probality Density Function
sns.distplot(titanic["age"],hist=False)
```

```python
# Plotting only Rugplot
sns.distplot(titanic["age"],hist = False, kde = False, rug=True)
```

### Where and How to use Distplot

```python
sns.distplot( titanic[titanic["survived"]== 1]["age"] )
sns.distplot( titanic[titanic["survived"]== 0]["age"] )
```
