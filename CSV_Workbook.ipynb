### Convertors

```python
pd.read_csv("./DataFolder/IPL Matches 2008-2020.csv").head()
```

```python
def Rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    elif name == "Mumbai Indians":
        return "MI"
    else:
        return name
```

```python
Rename("Royal Challengers Bangalore")
```

```python
Rename("Delhi Daredevils")
```

```python
pd.read_csv("./DataFolder/IPL Matches 2008-2020.csv", converters={"team1":Rename,"team2":Rename}).head()
```
