## Matplotlib Workbook

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Pie Chart

```python
Areas = ["Development","HR","Marketing","Sales","Customer services"]
Budget = [10,1.5,4,6,5]
```

```python
plt.pie(Budget, labels= Areas)
```

```python
plt.pie(Budget, labels= Areas)
plt.show()
```

```python
plt.pie(Budget, labels= Areas, autopct="%0.2f%%") # Add percentage to the pie chart
plt.show()
```

### Scatter Plot

```python
x = [23,45,12,50,85,11,40,36]
y= [35,46,22,28,78,21,35,47]

plt.scatter(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Rough Analysis")
```

### Plot

```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [233,456,770,120,222,445,200,320]

plt.plot(product_id,online_price, color="red",marker="*",markersize=10) # D for Diamond shape
plt.xlabel("product_id")
plt.ylabel("online_prices")
plt.title("price Comparision")
plt.show() # just displays the graph
```

```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [233,456,770,120,222,445,200,320]
offline_price = [275,510,900,150,275,500,250,390]
BlackMarket_price = [150,300,450,80,150,325,320,470]

plt.plot(product_id,online_price, color="red",marker="*",markersize=5,label = "Online Prices")
plt.plot(product_id,offline_price, color="blue",marker="D",markersize=5,label = "Offline Prices")
plt.plot(product_id,BlackMarket_price, color="green",marker="+",markersize=10,label = "BlackMarker Prices")

plt.legend() # To show the labels
plt.show()
```

### Bar Graphs

```python
year = [2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
Car_Price = [5000,5250,6000,5250,4500,8000,8670,7500,6750,9000]

plt.bar(year,Car_Price, color="red")
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("Price Vs Years")
```

```python
plt.barh(year,Car_Price, color="blue")
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("Price Vs Years")
```
