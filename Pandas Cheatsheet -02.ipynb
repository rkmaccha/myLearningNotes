```python
def runs_Scored(batsman_name):
    batsman_runsScored =  data[data['batsman'] == batsman_name]
    runs_S = batsman_runsScored.groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending=False).values[0]
    team_Against=batsman_runsScored.groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending=False).index[0]
    return (team_Against +" : " + str(runs_S) )

runs_Scored("RG Sharma")
```

```python
runs_Scored("CH Gayle")
```

### ISIN Function -> isin()

```python
# Find the most Destructive batsman in death overs in IPL
# Strike Rate = ( Number of runs / Number of ball played) * 100
# Condition - Minimum of 200 balls played in overs between 16-20
```

```python
data.shape
```

```python
data.groupby("batsman")['batsman_runs'].sum() # Gives number of runs scored
```

```python
data.groupby("batsman")[d'batsman_runs'].count() # Gives number of balls played
```

```python
data.columns
```

```python
data[data['over']>15].shape
```

```python
deathOvers = data[data['over']>15]
deathOvers.head()
```

```python
Batmans_Over200Balls =deathOvers.groupby("batsman")['batsman_runs'].count()
```

```python
type(Batmans_Over200Balls)
```

```python
Batmans_Over200Balls[Batmans_Over200Balls>200].head()
```

```python
BatsmanList = Batmans_Over200Balls[Batmans_Over200Balls>200].index.tolist()
```

```python
# Runs scored by the Batmans_Over200Balls
# Balls played by these Batsman

data[data['batsman'].isin(BatsmanList)]
```

```python
deathOvers[deathOvers['batsman'].isin(BatsmanList)]
```

```python
data[data['batsman'].isin(BatsmanList)].shape
```

```python
deathOvers[deathOvers['batsman'].isin(BatsmanList)].shape
```

```python
Final = data[data['batsman'].isin(BatsmanList)]
```

```python
Final = deathOvers[deathOvers['batsman'].isin(BatsmanList)]
```

```python
Final
```

```python
Runs = Final.groupby("batsman")["batsman_runs"].sum()
```

```python
Balls = Final.groupby("batsman")["batsman_runs"].count()
```

```python
strikeRate = (Runs/Balls) * 100
strikeRate.head()
```

```python
strikeRate.sort_values(ascending=False).head()
```

### MERGE

```python
# List of Orange Cap Holders
# Season - Batsman
```

```python
matches = pd.read_csv("iplmatches.csv")
matches.head()
```

```python
delivery = pd.read_csv("deliveries.csv")
delivery.head()
```

```python
delivery.merge(matches,left_on='match_id',right_on = 'id').shape
```

```python
new= delivery.merge(matches,left_on='match_id',right_on = 'id')
new.head()
```

```python
new.groupby(["season","batsman"])['batsman_runs'].sum()
```

```python
new.groupby(["season","batsman"])['batsman_runs'].sum().reset_index()
```

```python
new.groupby(["season","batsman"])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset="season",keep="first").sort_values("season")[["season","batsman"]]
```
